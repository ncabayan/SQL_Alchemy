```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#display the row's colums and data in dictionary format for measurement
first_row_measurement = session.query(Measurement).first()

first_row_measurement.__dict__
```

```python
#display the row's colums and data in dictionary format for station
first_row_station = session.query(Station).first()

first_row_station.__dict__
```

```python
for row in session.query(Measurement.date).limit(5):
    print(row)
```

```python
# get min and max dates and get the date from 12 months ago. 
min_date = engine.execute("select min(date) from Measurement;")
for record in min_date:
    print(record)

max_date = engine.execute("select max(date) from Measurement;")
for record in max_date:
    print(record)

twelve_months_ago = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(twelve_months_ago)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
#sort the results by date so that the data frame is already sorted by date
precipitation = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date >= twelve_months_ago).\
    order_by(Measurement.date).all()
```

```python
#put precipitation from last 12 months into data frame
precip_df = pd.DataFrame(precipitation)
precip_df.head()
```

```python
# reset the index to the date
precip_df.set_index('date').head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
precip_df.plot('date', 'prcp')
plt.xlabel("Date")
plt.ylabel("Precipitation")
plt.title("Precipitation Analysis For the Past Year")
plt.legend(["Precipitation"])
plt.tight_layout()
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
unique_stations = engine.execute("select count(distinct name) from Station;")
for record in unique_stations:
    print(record)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
most_active_stations= engine.execute("select station, count(station) as 'station_activity' from Measurement group by station order by station_activity desc;")
for record in most_active_stations:
    print(record)
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?

lowest_temp = engine.execute("select min(tobs) from Measurement where station like 'USC00519281';")
for record in lowest_temp:
    print(record)

highest_temp = engine.execute("select max(tobs) from Measurement where station like 'USC00519281';")
for record in highest_temp:
    print(record)

avg_temp = engine.execute("select avg(tobs) from Measurement where station like 'USC00519281';")
for record in avg_temp:
    print(record)

```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

# query to pull the last year of precipitation data for the busiest station
station_observation = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.station == 'USC00519281').\
    filter(Measurement.date > twelve_months_ago).\
    order_by(Measurement.date).all()
```

```python
#plot the histogram
station_observation_df=pd.DataFrame(station_observation)
plt.hist(station_observation_df['tobs'],12)
plt.xlabel("Recorded Temperature")
plt.ylabel("Number of Recorded Observations")
plt.title("Busiest Station Analysis For the Past Year")
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
#one way to count tmin, tavg, and tmax for dates 2012-02-28 to 2012-03-05
# for your trip using the previous year's data for those same dates.
t_min = engine.execute("select min(tobs) from Measurement where date between '2012-02-28' and '2012-03-05';")
for record in t_min:
    print(record)

t_avg = engine.execute("select avg(tobs) from Measurement where date between '2012-02-28' and '2012-03-05';")
for record in t_avg:
    print(record)

t_max = engine.execute("select max(tobs) from Measurement where date between '2012-02-28' and '2012-03-05';")
for record in t_max:
    print(record)
```
